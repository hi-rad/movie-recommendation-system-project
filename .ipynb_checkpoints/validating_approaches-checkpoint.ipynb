# Loading the Data

We will use the evaluation dataset we created (the models has not seen some ratings in this dataset). This dataset contains all movies and users presented in the training dataset, but for each user, some ratings are new and the train dataset does not contain them.

We will use those users not used in the training phase of the models (5850 users) for evaluations.

```python
if not os.path.exists(os.path.join(os.path.join(cwd, 'data', 'processed', 'evaluation_users.csv'))):
    all_users = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'users.csv'))

    users = []
    for (_, user) in all_users.iterrows():
        print(f'Checking User {user["id"]}')
        train_ratings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'train_user_ratings', f'{user["id"]}.csv'))
        evaluation_ratings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'evaluation_user_ratings', f'{user["id"]}.csv'))
        if len(evaluation_ratings[~evaluation_ratings['movieId'].isin(train_ratings['movieId'])]) > 0:
            users.append(user['id'])
    
    users_df = pd.DataFrame(users, columns=['id'])
    users_df.to_csv(os.path.join(cwd, 'data', 'processed', 'evaluation_users.csv'), index=False)

    del all_users
    del users_df
else:
    users_df = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'evaluation_users.csv'))
    users = users_df['id'].values
    
    del users_df

user_unseen_ratings = {}

for user in users:
    train_ratings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'train_user_ratings', f'{user}.csv'))
    evaluation_ratings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'evaluation_user_ratings', f'{user}.csv'))
    user_unseen_ratings[user] = evaluation_ratings[~evaluation_ratings['movieId'].isin(train_ratings['movieId'])]
```

```python
movie_actors_genre = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'movie_actors_genres.csv'))
average_movie_ratings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'movies_average_ratings.csv'))
movies_genres = {}
movies_casts = {}
movies_average_rating = {}
for (_, movie) in movie_actors_genre.iterrows():
    movies_genres[movie['movieId']] = movie['genres'].split('|')
    movies_casts[movie['movieId']] = movie['casts'].split('|')
for (_, movie) in average_movie_ratings.iterrows():
    movies_average_rating[movie['id']] = movie['rating']
movie_mappings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'movie_mappings.csv'))
train_average_movie_ratings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'train_average_movie_ratings.csv'))
genres = [' '.join(genre.split('_')[1:-1]).title() for genre in train_average_movie_ratings.drop(['userId'], axis=1).columns.tolist()]

del movie_actors_genre
del average_movie_ratings

clustering_predictions = {}
kd_tree_predictions = {}
autoencoder_predictions = {}
```

## KMeans Clustering

```python
clustering_data = train_average_movie_ratings[train_average_movie_ratings['userId'].isin(users)]
clustering_data_users_index = clustering_data['userId']
clustering_data = clustering_data.drop(['userId'], axis=1)

clustering_model = load(os.path.join(cwd, 'saved_models', 'k_means_model.pkl'))
cluster_for_all_trained_users = clustering_model.predict(train_average_movie_ratings.drop(['userId'], axis=1).values)
clusters_meta = {}
for index, cluster in enumerate(cluster_for_all_trained_users):
    if not cluster in clusters_meta:
        clusters_meta[cluster] = []
    clusters_meta[cluster].append(index)
del cluster_for_all_trained_users
predicted_clusters = clustering_model.predict(clustering_data.values)

cluster_genre_score = {}

for user_index, cluster in enumerate(predicted_clusters):
    clustering_predictions[clustering_data_users_index.iloc[user_index]] = {}
    if not cluster in cluster_genre_score:
        users_in_cluster = clusters_meta[cluster]
        users_in_cluster = train_average_movie_ratings[train_average_movie_ratings['userId'].isin(users_in_cluster)].drop(['userId'], axis=1)
        average_genre_ratings_in_cluster = {' '.join(k.split('_')[1:-1]).title():v for k, v in dict(users_in_cluster.mean(axis=0)).items()}
        average_genre_ratings_in_cluster = {k: (v/10) for k, v in average_genre_ratings_in_cluster.items()}
        cluster_genre_score[cluster] = average_genre_ratings_in_cluster
    for (_, movie) in user_unseen_ratings[clustering_data_users_index.iloc[user_index]].iterrows():
        movie_genres = movies_genres[int(movie['movieId'])]
        movie_score = sum([score for genre, score in cluster_genre_score[cluster].items() if genre in movie_genres])
        predicted_rating = movies_average_rating[int(movie['movieId'])] * movie_score
        actual_rating = movie['rating']
        clustering_predictions[clustering_data_users_index.iloc[user_index]][int(movie['movieId'])] = {
            'actual_rating': actual_rating,
            'predicted_rating': predicted_rating,
            'movie_average_rating': movies_average_rating[int(movie['movieId'])],
            'clustering_score': movie_score,
        }

print("KMeans Clustering Done!")
```

## KD-Tree

```python
encoded_movies_actors_genres_data = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'genre_cast_encoded_movies.csv'))
movies_index = encoded_movies_actors_genres_data['movieId']
total_number_of_movies = len(encoded_movies_actors_genres_data)

tree = load(os.path.join(cwd, 'saved_models', 'kd_tree_model.pkl'))

movies_distance_score = {}
calculated_movies = []

print('Creating Movie Distances For Each User')
for user in users:
    print(f'Processing User {user}')
    user_unseen_movies = user_unseen_ratings[user]
    user_unseen_movies = user_unseen_movies[~user_unseen_movies['movieId'].isin(calculated_movies)]
    user_unseen_movies = user_unseen_movies['movieId'].unique().tolist()
    if len(user_unseen_movies):
        tree_data = encoded_movies_actors_genres_data[encoded_movies_actors_genres_data['movieId'].isin(user_unseen_movies)]
        tree_data_movies_index = tree_data['movieId']
        tree_data = tree_data.drop(['imdbId', 'tmdbId', 'movieId', 'genres', 'casts'], axis=1)
        tree_data = tree_data.values

        dist, ind = tree.query(tree_data, k=total_number_of_movies)

        for source_movie_index, distances in enumerate(dist):
            source_movie_id = tree_data_movies_index.iloc[source_movie_index]
            movies_distance_score[source_movie_id] = {}
            calculated_movies.append(source_movie_id)
            maximum_distance = max(distances)
            for index_in_results, distance in enumerate(distances):
                target_movie_index = ind[source_movie_index][index_in_results]
                target_movie_id = movies_index.iloc[target_movie_index]
                movie_distance = distance
                # A distance of zero means a hit, so we use (1 - distance/maximum_distance) to calculate the distance score
                movies_distance_score[source_movie_id][target_movie_id] = 1 - (movie_distance/maximum_distance)

    train_ratings = pd.read_csv(os.path.join(cwd, 'data', 'processed', 'train_user_ratings', f'{user}.csv'))
    kd_tree_predictions[user] = {}
    for (_, movie) in user_unseen_ratings[user].iterrows():
        distance_based_rating_sum = 0
        sum_of_distances = 0
        for (__, seen_movie) in train_ratings.iterrows():
            distance_based_rating_sum += (movies_distance_score[movie['movieId']][seen_movie['movieId']] * movies_average_rating[int(movie['movieId'])])
            sum_of_distances += movies_distance_score[movie['movieId']][seen_movie['movieId']]
        predicted_rating = distance_based_rating_sum / sum_of_distances
        actual_rating = movie['rating']
        kd_tree_predictions[user][int(movie['movieId'])] = {
            'actual_rating': actual_rating,
            'predicted_rating': predicted_rating,
        }

print("KD-Tree Done!")
```

## Autoencoder

```python
from models import AutoEncoder
import torch

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = AutoEncoder(num_features=len(movie_mappings), dimensions=[400, 200, 400])
model = model.to(device)
model.load_state_dict(torch.load(os.path.join(cwd, 'saved_models', 'final_model_400_200_400.pth')))
```

We load the data used for training and for evaluations, we check the predicted value of the model for those items available in the evaluation ratings file (previously loaded in the `user_unseen_ratings` variable).

```python
from data_loader import Dataset
from torch.utils.data import DataLoader

movie_mappings_dict = {movie_info['rating_id']:movie_info['id'] for (_, movie_info) in movie_mappings.iterrows()}

evaluation_dat = Dataset(os.path.join(cwd, 'data', 'processed'), 'train_user_ratings', users, movie_mappings_dict)
batch_size = 128
evaluation_dl = DataLoader(dataset=evaluation_dat, batch_size=batch_size, shuffle=False, num_workers=1)

batch_number = 0
for i, data in enumerate(evaluation_dl, 0):
    batch_number += 1
    model.eval()
    inputs = data
    inputs = inputs.to(device)

    inputs = inputs.float()

    outputs = model(inputs)
    user_number_in_batch = 0
    for output in outputs:
        user_number_in_batch += 1
        predicted_values = output.detach().cpu().numpy()
        user_id = users[((batch_number - 1) * batch_size) + (user_number_in_batch - 1)]
        user_true_ratings = user_unseen_ratings[user_id]
        autoencoder_predictions[user_id] = {}
        for (_, movie) in user_unseen_ratings[user_id].iterrows():
            autoencoder_predictions[user_id][int(movie['movieId'])] = {
                'actual_rating': movie['rating'],
                'predicted_rating': predicted_values[movie_mappings_dict[movie['movieId']]],
            }
print("Autoencoder Done!")
```

## Hybrid Approach
In this method, we want to combine the Autoencoder approach with the KD-Tree approach so that we have both a content-based and a collaborative filtering based method. To do so, we will use 0.7 as coefficient for the collaborative filtering rating, and 0.3 for the content-based filtering method (Sum of the coefficients sum is 1). For each movie, the rating produced by each method will be multiplied by its corresponding method's coefficient and summed together.

```python
hybrid_predictions = {}

for user in users:
    hybrid_predictions[user] = {}
    kd_tree_user_predictions = kd_tree_predictions[user]
    autoencoder_user_predictions = autoencoder_predictions[user]
    for (movie_id, rating_values) in kd_tree_user_predictions.items():
        kd_value = kd_tree_user_predictions[movie_id]['predicted_rating']
        ae_value = autoencoder_user_predictions[movie_id]['predicted_rating']
        hybrid_predictions[user][movie_id] = {
                'actual_rating': rating_values['actual_rating'],
                'predicted_rating': (ae_value * 0.7) + (kd_value * 0.3)
            }

print("Hybrid Done!")
```

# Evaluations

```python
from sklearn.metrics import mean_squared_error, accuracy_score, precision_score, recall_score, f1_score

all_clustering_predicted_ratings = []
all_clustering_actual_ratings = []
all_kd_tree_predicted_ratings = []
all_kd_tree_actual_ratings = []
all_autoencoder_predicted_ratings = []
all_autoencoder_actual_ratings = []
all_hybrid_predicted_ratings = []
all_hybrid_actual_ratings = []
for user_id in users:
    kd_tree_user_predictions = kd_tree_predictions[user_id]
    for (movie_id, rating_values) in kd_tree_user_predictions.items():
        all_clustering_predicted_ratings.append(clustering_predictions[user_id][int(movie_id)]['predicted_rating'])
        all_clustering_actual_ratings.append(clustering_predictions[user_id][int(movie_id)]['actual_rating'])
        all_kd_tree_predicted_ratings.append(kd_tree_predictions[user_id][int(movie_id)]['predicted_rating'])
        all_kd_tree_actual_ratings.append(kd_tree_predictions[user_id][int(movie_id)]['actual_rating'])
        all_autoencoder_predicted_ratings.append(autoencoder_predictions[user_id][int(movie_id)]['predicted_rating'])
        all_autoencoder_actual_ratings.append(autoencoder_predictions[user_id][int(movie_id)]['actual_rating'])
        all_hybrid_predicted_ratings.append(hybrid_predictions[user_id][int(movie_id)]['predicted_rating'])
        all_hybrid_actual_ratings.append(hybrid_predictions[user_id][int(movie_id)]['actual_rating'])
mse = {
    'clustering': mean_squared_error(all_clustering_actual_ratings, all_clustering_predicted_ratings),
    'kd_tree': mean_squared_error(all_kd_tree_actual_ratings, all_kd_tree_predicted_ratings),
    'autoencoder': mean_squared_error(all_autoencoder_actual_ratings, all_autoencoder_predicted_ratings),
    'hybrid': mean_squared_error(all_hybrid_actual_ratings, all_hybrid_predicted_ratings),
}
cl_true_classes = [1 if rating >= 3.5 else 0 for rating in all_clustering_actual_ratings]
cl_pred_classes = [1 if rating >= 3.5 else 0 for rating in all_clustering_predicted_ratings]
kd_true_classes = [1 if rating >= 3.5 else 0 for rating in all_kd_tree_actual_ratings]
kd_pred_classes = [1 if rating >= 3.5 else 0 for rating in all_kd_tree_predicted_ratings]
ae_true_classes = [1 if rating >= 3.5 else 0 for rating in all_autoencoder_actual_ratings]
ae_pred_classes = [1 if rating >= 3.5 else 0 for rating in all_autoencoder_predicted_ratings]
hb_true_classes = [1 if rating >= 3.5 else 0 for rating in all_hybrid_actual_ratings]
hb_pred_classes = [1 if rating >= 3.5 else 0 for rating in all_hybrid_predicted_ratings]
accuracy = {
    'clustering': accuracy_score(cl_true_classes, cl_pred_classes),
    'kd_tree': accuracy_score(kd_true_classes, kd_pred_classes),
    'autoencoder': accuracy_score(ae_true_classes, ae_pred_classes),
    'hybrid': accuracy_score(hb_true_classes, hb_pred_classes),
}
precision = {
    'clustering': precision_score(cl_true_classes, cl_pred_classes),
    'kd_tree': precision_score(kd_true_classes, kd_pred_classes),
    'autoencoder': precision_score(ae_true_classes, ae_pred_classes),
    'hybrid': precision_score(hb_true_classes, hb_pred_classes),
}
recall = {
    'clustering': recall_score(cl_true_classes, cl_pred_classes),
    'kd_tree': recall_score(kd_true_classes, kd_pred_classes),
    'autoencoder': recall_score(ae_true_classes, ae_pred_classes),
    'hybrid': recall_score(hb_true_classes, hb_pred_classes),
}
f1_measure = {
    'clustering': f1_score(cl_true_classes, cl_pred_classes),
    'kd_tree': f1_score(kd_true_classes, kd_pred_classes),
    'autoencoder': f1_score(ae_true_classes, ae_pred_classes),
    'hybrid': f1_score(hb_true_classes, hb_pred_classes),
}

print("Calculations Done!")
```

```python
from beautifultable import BeautifulTable

headers = ['Clustering', 'KD_Tree', 'Autoencoder', 'Hybrid']

table = BeautifulTable()
for header in headers:
    table.rows.append([mse[header.lower()], accuracy[header.lower()], precision[header.lower()], recall[header.lower()], f1_measure[header.lower()]])
table.columns.header = ["MSE", "Accuracy", "Precision", "Recall", "F1 Measure"]
table.rows.header = [' '.join(header.split('_')) for header in headers]
print(table)
```

```python
labels = ['MSE', 'Accuracy', 'Precision', 'Recall', 'F1 Measure']
clustering_res = [mse['clustering'], accuracy['clustering'], precision['clustering'], recall['clustering'], f1_measure['clustering']]
kd_res = [mse['kd_tree'], accuracy['kd_tree'], precision['kd_tree'], recall['kd_tree'], f1_measure['kd_tree']]
ae_res = [mse['autoencoder'], accuracy['autoencoder'], precision['autoencoder'], recall['autoencoder'], f1_measure['autoencoder']]
hybrid_res = [mse['hybrid'], accuracy['hybrid'], precision['hybrid'], recall['hybrid'], f1_measure['hybrid']]

clustering_res = [float(f"{res:.4f}") for res in clustering_res]
kd_res = [float(f"{res:.4f}") for res in kd_res]
ae_res = [float(f"{res:.4f}") for res in ae_res]
hybrid_res = [float(f"{res:.4f}") for res in hybrid_res]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


plt.rcParams["figure.figsize"]=(18, 5)
fig, ax = plt.subplots()
rects1 = ax.bar(x - (3*width/4), clustering_res, width/2, label='Clustering')
rects2 = ax.bar(x - width/4, kd_res, width/2, label='KD Tree')
rects3 = ax.bar(x + width/4, ae_res, width/2, label='Autoencoder')
rects4 = ax.bar(x + (3*width/4), hybrid_res, width/2, label='Hybrid')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Metrics')
ax.set_title('Metrics Grouped by Method')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)

fig.tight_layout()

plt.savefig(os.path.join(cwd, 'methods_comparison.png'))
```

# Conclusion

As can be seen by the above table and chart, the Hybrid approach has the lowest Mean Squared Error and the highest Accuracy, Precision, and Recall. Therefore, the hybrid approach would be a better option.

The KD-Tree approach is only based on the genres and casts. It is a content-based only approach. The Clustering approach takes the ratings and genere into account. As a result, it is using both the collaborative filtering and content-based filtering approaches. However, it fails to produce more accurate results compared to the other methods. The autoencoder is a collaborative filtering only approach. The input is the user ratings. The movies use has not yet rated will be set to zero. Then the model tries to recreate the input ratings in the output. In the process of learning, the weights will be updated based on the training users ratings. As a result, the optimization of the network's weights is done using the collaboration of the previous users' ratings. When the model generates the outputs, it also predicts ratings for the unseen movies and these ratings will be used as predictions.

Finally, the hybrid approach uses both the collaborative filtering approach (the autoencoder) and the content-based filtering approach (KD-Tree). We have assigned a coefficient of 0.7 for the autoencoder results and a coefficient of 0.3 for the KD-Tree results, and added the final values together. The experiments show that the final Hybrid results are better than the other three approaches.
